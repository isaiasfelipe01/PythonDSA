# 2 - MANIPULAÇÃO DE ARQUIVOS TXT

```python
texto = "Cientista de Dados pode ser uma excelente alternativa de carreira.\n"
#duas formas de adicionar str a variável texto
texto = texto + "Esses profissionais precisam saber como programar em python.\n"
texto += "E, claro, devem ser proficientes em Data Science."
print(texto)
```

```python
import os
#criando arquivo para gravação
arquivo = open(os.path.join('arquivos/cientista.txt'), 'w')
for palavra in texto.split():
    arquivo.write(palavra+' ')
arquivo.close()
arquivo = open('arquivos/cientista.txt', 'r')
conteudo = arquivo.read()
arquivo.close()
print(len(conteudo))
print(conteudo)
```

```python
#usando o metodo with não precisa usar o método close
#manipulando arquivos txt
with open('arquivos/cientista.txt', 'w') as arquivo:
    arquivo.write(texto[:19])
    arquivo.write('\n')
    arquivo.write(texto[28:66])

with open('arquivos/cientista.txt', 'r') as arquivo:
    conteudo = arquivo.read()
print(len(conteudo))
print(conteudo)
```

```python
#manipulando arquivos csv
import csv
with open('arquivos/numeros.csv', 'w') as arquivo:
    #Criando objeto de gravaão
    write = csv.writer(arquivo)

    #gravando no arquivo linha a linha
    write.writerow(('nota1', 'nota2', 'nota3'))
    write.writerow((63,87,92))
    write.writerow((61,79,76))
    write.writerow((72,64,91))
```

```python
#leitura de arquivos csv
with open('arquivos/numeros.csv', 'r', encoding='utf8', newline = '\r\n') as arquivo:
    #Cria um objeto de leitura
    leitor = csv.reader(arquivo)
    lista = []
    for x in leitor:
        lista.append(x)
        print(x)
print(lista[1:])
for linha in lista[1:]:
    print(linha)
```

```python
#gerando uma lista com dados do arquivo csv
with open('arquivos/numeros.csv', 'r') as arquivo:
    leitor = csv.reader(arquivo)
    dados = list(leitor)
print(dados)
for linha in dados[1:]:
    print(linha)
```

# 3 - MANIPULAÇÃO DE ARQUIVO JSON

```python
#Criando dicionário
dict_guido = {
    'nome': 'Guido Van Rossum',
    'linguagem': 'Python',
    'similar': ['c', 'Modular-3', 'lisp'],
    'users': 1000000
}
for k,v in dict_guido.items():
    print(k,v)

import json
#função dumps para converter dicionário em JSON
json.dumps(dict_guido)
#criando aquivo JSON
with open('arquivos/dados.json', 'w') as arquivo:
    arquivo.write(json.dumps(dict_guido))
#lendo arquivo JSON
with open('arquivos/dados.json', 'r') as arquivo:
    texto = arquivo.read()
    #carregando arquivo JSON
    dados = json.loads(texto)
print(dados)
print(dados['nome'])
```

```python
#imprimindo um arquio JSON copiado da internet
from urllib.request import urlopen
response = urlopen("https://vimeo.com/api/v2/video/57733101.json").read().decode('utf8')
dados = json.loads(response)[0]
#imprimindo conteúdo extraido da web
dados
```

```python
#extraindo dados do arquivo JSON da web
print('Título: ', dados['title'])
print('URL: ', dados['url'])
print('duração: ', dados['duration'])
print('Números de Visualizações: ', dados['stats_number_of_plays'])
```

### Copiando conteúdo de um arquivo para outro

# 5 - FUNÇÕES

### MAP

```python
def potencia(x):
    return x ** 2
#Elevando ao quadrado todos os números de uma lista
numeros = [1, 2, 3, 4, 5]
num_ao_quadrado = list(map(potencia, numeros))
print(num_ao_quadrado)
##Função para converter temperaturas
def tempc(t):
    return float((t-32)*5/9)
def tempf(t):
    return float((t*9/5)+32)
t = [0, 22.5, 40, 100]
for temp in map(tempf, t):
    print(temp)

#usando a função lambda 
list(map(lambda x: (5.0/9)*(x-32), t))

#usando função lamba para somar listas
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]
print('Soma de listas com lambda: ', list(map(lambda x, y: x+y, a, b)))

c = [9, 10, 11, 12]
print('Soma de 3 listas com lambda: ', list(map(lambda x, y, z: x + y + z , a, b, c)))
```

### Função Reduce

```python
from functools import reduce

lista = [47, 11, 42, 13]
print(lista)
def soma(a, b):
    x = a + b
    return x
#função para somar valores da lista
print(reduce(soma, lista))
print(reduce(lambda x, y: x + y, lista))

#função para encontrar o maior valor
max_find = lambda x, y: x if x > y else y
print(reduce(max_find, lista))
```

### Função Filter

```python
print(lista)
#filtrando valores pares da lista 
#método 1
verif_par = (lambda x: True if x % 2 == 0 else False)
print(list(filter(verif_par, lista)))
#método 2
print(list(filter(lambda x: x%2 == 0, lista)))
#filtrando valores maiores que 13
print(list(filter(lambda x: x>13, lista)))
```

### Função Zip

```python
x = [1, 2, 3]
y = [4, 5, 6]
#Combinando listas
print(list(zip(x, y)))
#Se a função não consegue combinar ela descarta
print(list(zip('ABCD', 'xy')))
a = [1, 2, 3]
b = [4, 5, 6, 7, 8]
print(list(zip(a, b)))
d1 = {'a':1, 'b':2}
d2 = {'c':4, 'd':5}
print(list(zip(d1, d2)))
print(list(zip(d1, d2.values())))
#função para trocar valores de dicionários
def trocaValores(d1, d2):
    dicTemp = {}
    for d1key, d2val in zip(d1, d2.values()):
        dicTemp[d1key] = d2val
    return dicTemp
trocaValores(d1, d2)
troca_val = lambda d1, d2: {k: d2[v] for k, v in zip(d1, d2)}
troca_val(d1, d2)
```

### Função Enumerate

```python
seq = ['a', 'b', 'c']
list(enumerate(seq))
for indice, valor in enumerate(seq):
    print(indice, valor)
print('---------')
for indice, valor in enumerate(seq):
    if indice >= 2:
        break
    print(indice, valor)
print('---------')
lista = ['Marketing', 'Tecnologia', 'Business']
for i, valor in enumerate(lista):
    print(i, valor)
```

# 6 - ERROS E EXCESSÕES

### TRY, EXCEPT, FINALLY

```python
#Usando try e except
try:
    8 + 's'
except TypeError:
    print('Operação não permitida')
```

```python
try:
    f = open('arquivos/testandoerros.txt', 'w')
    f.write('Gravando no arquivo')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser salvo.')
else:
    print('Conteúdo gravado com sucesso.')
    f.close()
```

```python
try:
    f = open('arquivos/testandoerros', 'r')
    f.write('Gravando no arquivo')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser lido.')
else:
    print('Conteúdo gravado com sucesso.')
    f.close()
```

```python
try:
    f = open('arquivos/testandoerros.txt', 'w')
    f.write('Gravando no arquivo')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser salvo.')
else:
    print('Conteúdo gravado com sucesso.')
    f.close()
#operação independente de ter erro ou não
finally:
    print('Comandos no bloco finally são sempre executados.')
```

## Tratando possibilidades de erro

```python
def askint():
        try:
            val = int(input('Digite um número: '))
        except:
            print('Você não digitou um número!')
        finally:
            print('Obrigado!')
askint()
```

```python
def askint():
        try:
            val = int(input('Digite um número: '))
        except:
            print('Você não digitou um número!')
            val = int(input('Tente novamente. Digite um número: '))
        finally:
            print('Obrigado!')
askint()
```

```python
def askint():
        while True:
            try:
                val = int(input('Digite um número: '))
            except:
                print('Você não digitou um número!')
                continue
            else:
                print('Obrigado por digitar um número!')
                break
            finally:
                print('Fim da execução!')
        print(val)
askint()
```
